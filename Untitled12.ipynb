```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from arch import arch_model
import yfinance as yf
from datetime import timedelta
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import mean_absolute_percentage_error as mape
import warnings
warnings.filterwarnings("ignore")
```

```python
def download_day_data(instrument='BTC-USD', start='2020-01-01', end='2024-01-01'):
  return yf.download(instrument, start=start, end=end)

def download_hour_data(instrument='BTC-USD', start='2023-01-01', end='2024-01-01'):
   return yf.download(instrument, interval='1h', start=start, end=end)

def calculate_returns(data, is_dropna=True):
  returns = data['Adj Close'].pct_change()
  return returns.dropna() if is_dropna else returns

def calculate_volatility(data, window, is_percentage=True, is_dropna=True):
  returns = calculate_returns(data, is_dropna=is_dropna)
  volatility = returns.rolling(window=window).std().dropna() if is_dropna else returns.rolling(window=window).std()
  return volatility.to_numpy() * (100 if is_percentage else 1)

def calculate_day_volatility(data, is_percentage=True, is_dropna=True):
  hour_data = data
  hour_data['Hour'] = data.index.hour
  hour_data['Date'] = data.index.date
  hour_data['Returns'] = hour_data['Close'].pct_change()
  volatility = hour_data.groupby('Date')['Returns'].agg(lambda x: x.std() * np.sqrt(24.0))
  return volatility.to_numpy() * (100 if is_percentage else 1)


def plot_volatility(data, window):
  volatility = calculate_volatility(data, window, is_persentate=True)
  plt.figure(figsize=(14, 7))
  plt.plot(volatility, label=f'{window}-дневная волатильность', color='blue')
  plt.title('Волатильность актива')
  plt.xlabel('Дата')
  plt.ylabel('Волатильность (%)')
  plt.legend()
  plt.grid()
  plt.show()

def fit_garch(data, model_name='Garch', p=1, q=1, is_print_summary=True):
  model = arch_model(calculate_returns(data), vol=model_name, p=p, q=q)
  garch_fit = model.fit(disp="off")
  forecast = garch_fit.forecast(start=0)
  predicted_volatility = np.sqrt(forecast.variance.values)

  if is_print_summary:
    print(garch_fit.summary())

  return predicted_volatility.flatten()

def fit_naive_model(volatility):
  return volatility[:-1]

def plot_forecast_vs_realize(data, realize_volatility, predicted_volatility):
  plt.figure(figsize=(12, 6))
  size = min(realize_volatility.shape[0], predicted_volatility.shape[0])

  plt.plot(data.index[-size:], realize_volatility[-size:], label=f'Real Volatility', color='blue')
  plt.plot(data.index[-size:], predicted_volatility[-size:], label='Predicted Volatility', color='red', linestyle='-')
  plt.title('Real vs Predicted Volatility of Bitcoin')
  plt.xlabel('Date')
  plt.ylabel('Volatility')
  plt.legend()
  plt.grid()
  plt.show()

def calculate_metrics(realize_value, predicted_value):
  size = min(realize_value.shape[0], predicted_value.shape[0])
  realize_value = realize_value[-size:]
  predicted_value = predicted_value[-size:]
  return (mse(realize_value, predicted_value), mae(realize_value, predicted_value))
```

```python
data = download_day_data('BTC-USD', start='2020-01-01', end='2024-01-01')
window = 7
garch_forecast = fit_garch(data=data, p=1, q=1)
realize_volatility = calculate_volatility(data=data, window=window, is_percentage=False)
plot_forecast_vs_realize(data=data, realize_volatility=realize_volatility, predicted_volatility=garch_forecast)

naive_forecast = fit_naive_model(realize_volatility)
print('garch:', calculate_metrics(realize_volatility, garch_forecast))
print('naive:', calculate_metrics(realize_volatility, naive_forecast))
```

```python
def plot_table(data):
  table = pd.DataFrame(data, columns=["Model", "MSE", "MAE"])
  plt.figure(figsize=(8, 4))
  sns.set(style="whitegrid")
  styled_table = table.style.background_gradient(cmap="Oranges", subset=["MSE", "MAE"])
  styled_table = styled_table.set_properties(**{'text-align': 'center'})
  display(styled_table)

def grid_fit_model(realize_volatility, day_data, model='Garch',mse_scaler=1, mae_scaler=1, p_start=1, p_end=1, q_start=1, q_end=1):
  table = []
  table.append(['naive', 1, 1])

  for p in range(p_start, p_end+1):
    for q in range(q_start, q_end+1):
      garch_forecast = fit_garch(day_data, model_name=model, p=p, q=q, is_print_summary=False)
      metrics = calculate_metrics(realize_volatility, garch_forecast)

      table.append([f'{model}(p={p}, q={q})', mse_scaler/metrics[0], mae_scaler/metrics[1]])

  return table
```

```python
naive_model_metrics = calculate_metrics(realize_volatility, naive_forecast)
naive_mse = naive_model_metrics[0]
naive_mae = naive_model_metrics[1]

garch_table = grid_fit_model(realize_volatility=realize_volatility, day_data=day_data, model='Garch',mse_scaler=naive_mse, mae_scaler=naive_mae, p_start=1, p_end=5, q_start=1, q_end=5)
plot_table(garch_table)
```
